```python
# Check for missing values in each column
missing_values = p1.isnull().sum()

# Calculate the percentage of missing values
missing_percentage = (missing_values / len(p1)) * 100

print("Columns with Missing Values and Their Percentages:")
print(pd.DataFrame({'Missing Values': missing_values[missing_values > 0], 'Percentage': missing_percentage[missing_percentage > 0]}))
```

```python
# Function to identify and count outliers using IQR method
def count_outliers(column):
    # Calculating the IQR for the column
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1

    # Defining the lower and upper bounds to identify outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identifying and counting outliers
    outliers = column[(column < lower_bound) | (column > upper_bound)]
    return len(outliers)

# Creating a list to store the results
outliers_count_list = []

# Iterating over columns in the DataFrame
for column in p1.columns:
    # Checking if the column contains numerical data
    if pd.api.types.is_numeric_dtype(p1[column]):
        # Counting outliers for the current column
        outliers_count = count_outliers(p1[column])
        
        # Calculating the total number of data points in the column
        total_data_points = len(p1[column])
        
        # Calculating the percentage of outliers
        percentage_outliers = (outliers_count / total_data_points) * 100
        
        # Appending the results to the list
        outliers_count_list.append({'Column': column, 'Outliers Count': outliers_count, 'Percentage Outliers': percentage_outliers})

# Creating a DataFrame from the list of results
outliers_count_df = pd.DataFrame(outliers_count_list)

print(outliers_count_df)
```

> **Data cleaning:**

```python
# List of columns to drop
columns_to_drop = ['neighbourhood_group_cleansed', 'bathrooms', 'calendar_updated', 'license']

# Drop the specified columns
p1_cleaned = p1.drop(columns=columns_to_drop, axis=1)
```

```python
columns_na = ['host_response_rate', 'host_acceptance_rate']

# Iterating over the columns and replacing 'N/A' with the mean
for column in columns_na:
    
    # Converting the columns to numeric type, coercing errors so that 'N/A' is treated as NaN
    p1_cleaned[column] = pd.to_numeric(p1_cleaned[column], errors='coerce')
    
    # Calculating the mean of the column (ignoring NaN)
    mean_column = np.nanmean(p1_cleaned[column])
    
    # Replacing 'N/A' with the mean of the column
    p1_cleaned[column].replace(np.nan, mean_column, inplace=True)

```

```python
missing_values = p1_cleaned.isnull().sum()

missing_percentage = (missing_values / len(p1_cleaned)) * 100

print("Columns with Missing Values and Their Percentages:")
print(pd.DataFrame({'Missing Values': missing_values[missing_values > 0], 'Percentage': missing_percentage[missing_percentage > 0]}))
```

```python
# Define the columns with outliers
outlier_columns = ['review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness',
                   'review_scores_checkin', 'review_scores_communication', 'review_scores_location',
                   'review_scores_value', 'accommodates']

# Impute outliers with the median
for column in outlier_columns:
    median_value = p1_cleaned[column].median()
    p1_cleaned[column] = np.where(
        (p1_cleaned[column] < p1_cleaned[column].quantile(0.05)) | (p1_cleaned[column] > p1_cleaned[column].quantile(0.95)),
        median_value,
        p1_cleaned[column]
    )
```
